# baseline model

```python
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

# Prepare X and y
X = data[['neu','ext','ope','agr','con']].fillna(0).round(2)
y = data['PSS10_avg'].fillna(0)

baseline_model = DummyRegressor(strategy="median") # Baseline model that predicts "median"

cv_results = cross_validate(baseline_model,X,y) # cross validate baseline

cv_results['test_score'].mean()
```

# linear model 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split


df=data[['neu','ext','ope','agr','con','PSS10_avg']].dropna()

y=df['PSS10_avg']
X=df[['neu','ext','ope','agr','con']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)

model=LinearRegression()
results=cross_validate(model,X_train,y_train,cv=2,scoring='r2')
results['test_score'].mean()
model.fit(X_train,y_train)
#model.coef_
#model.intercept_
#model.predict([[1,2,3,4,5]])

predicted_Val=model.predict(X_test)[0]
predicted_Val
results['test_score'].mean()
```
